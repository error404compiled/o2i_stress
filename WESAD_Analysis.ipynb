# WESAD Dataset Analysis

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("/home/deepin/Desktop/projects/hrv/WESAD/WESAD/S3chest.csv")
```

```python
df = df[df['ID'] == 2]
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df['label'].unique()
```

```python
df = df[df['label'] < 4]
df = df[df['label'] !=0]
df_1=np.unique(df['label'])
df_1
```

```python
sns.set_context("paper", rc={"lines.linewidth": 2.5})
sns.set_palette("binary_d")
```

```python
# Neutral
sns.lineplot(data=df[df['label'] == 1].reset_index(drop=True)['chestResp'])
plt.ylabel('RESP')
plt.xlabel('Sequential Data-Points')
plt.show()
```
